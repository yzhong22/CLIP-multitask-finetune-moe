```python
import os
import glob
from tqdm import tqdm
import numpy as np
import pandas as pd
```

```python
root = "/media/yesindeed/DATADRIVE1/mount/remote_cse/datasets/multitask-moe/CheXpert-v1.0-small/"

df = pd.read_csv(f"{root}/train.csv")

df
```

```python
df["Edema"].replace(-1, 1, inplace=True)
df["Edema"].fillna(0, inplace=True)

df["Atelectasis"].replace(-1, 1, inplace=True)
df["Atelectasis"].fillna(0, inplace=True)

df["Cardiomegaly"].replace(-1, 0, inplace=True)
df["Cardiomegaly"].fillna(0, inplace=True)

df["Consolidation"].replace(-1, 0, inplace=True)
df["Consolidation"].fillna(0, inplace=True)

df["Pleural Effusion"].replace(-1, 0, inplace=True)
df["Pleural Effusion"].fillna(0, inplace=True)

df["Pneumonia"].replace(-1, 0, inplace=True)
df["Pneumonia"].fillna(0, inplace=True)

df["Pneumothorax"].replace(-1, 0, inplace=True)
df["Pneumothorax"].fillna(0, inplace=True)

df["No Finding"].replace(-1, 0, inplace=True)
df["No Finding"].fillna(0, inplace=True)

df
```

```python
df_split = pd.read_csv(f"{root}/metadata_multilabel.csv")

test_imgs = df_split.loc[df_split["split"] == "test"]["id"].tolist()
```

```python
dict_out = {"id": [], "exam_id": [], "path": [], "class": [], "exam_class": [], "patient": []}

target_diseases = [
    "Cardiomegaly",
    "Edema",
    "Consolidation",
    "Atelectasis",
    "Pleural Effusion",
    "Pneumonia",
    "Pneumothorax",
]


df_out_all = []

for disease in target_diseases:
    df_sub = df.loc[(df[disease] == 1) | (df["No Finding"] == 1)]

    dict_out = {"id": [], "exam_id": [], "path": [], "class": [], "exam_class": [], "patient": []}

    for i in range(len(df_sub)):
        item = df_sub.iloc[i]

        path = item["Path"]
        img_path = path.replace("CheXpert-v1.0-small/", "./")

        assert os.path.exists(f"{root}/{img_path.replace('./', '')}")

        patient, study, img = path.split("/")[-3:]

        if item["No Finding"] == 1:
            cls = "no findings"
        else:
            cls = disease.lower()

        dict_out["id"].append(f"{patient}/{study}/{img.replace('.jpg', '')}")
        dict_out["exam_id"].append(f"{patient}/{study}/{img.replace('.jpg', '')}")

        dict_out["path"].append(img_path)
        dict_out["patient"].append(patient)

        dict_out["class"].append(cls)
        dict_out["exam_class"].append(cls)

    df_out = pd.DataFrame.from_dict(dict_out)

    df_out["dataset"] = f"chexpert-{disease.lower()}"
    df_out["class_num"] = 2
    df_out["contains_negative"] = 1
    df_out["modality"] = "xray"

    df_out = df_out.sample(frac=1).reset_index(drop=True)

    df_out["split"] = "train"

    df_out.loc[df_out["id"].isin(test_imgs), "split"] = "test"

    df_out_all.append(df_out)

df_out_all = pd.concat(df_out_all, axis=0)
df_out_all = df_out_all.drop("patient", axis=1)
```

```python
df_out_all.to_csv(f"{root}/metadata.csv", index=False)
```
