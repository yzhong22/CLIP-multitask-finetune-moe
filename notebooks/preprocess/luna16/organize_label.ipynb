```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
import glob
```

```python
root = "/media/yesindeed/DATADRIVE1/mount/remote_cse/datasets/multitask-moe/luna16"
```

```python
df_annotation_2d = pd.read_csv(os.path.join(root, "annotations_2d.csv"))
```

```python
df_annotation_2d
```

```python
dict_out = {"id": [], "exam_id": [], "path": [], "class": [], "exam_class": []}

for id in df_annotation_2d["id"].unique():
    sub_df = df_annotation_2d.loc[df_annotation_2d["id"] == id]

    is_node = np.max(sub_df["is_node"])

    assert len(sub_df["path"].unique()) == 1

    path = sub_df["path"].tolist()[0].replace(
        "/research/d5/gds/yzhong22/datasets/multitask-moe/luna16/", "./")

    assert os.path.exists(os.path.join(root, path)), f"{path}"

    cls = "node" if is_node == 1 else "no findings"

    dict_out["id"].append(id)
    dict_out["exam_id"].append(id)
    dict_out["path"].append(path)

    dict_out["class"].append(cls)
    dict_out["exam_class"].append(cls)

df_out = pd.DataFrame.from_dict(dict_out)
```

```python
import numpy as np

df_out["dataset"] = "luna16"
df_out["class_num"] = 1
df_out["contains_negative"] = 1
df_out["modality"] = "ct"

df_out["split"] = "test"
```

```python
df_out.to_csv(f"{root}/metadata.csv", index=False)
```
