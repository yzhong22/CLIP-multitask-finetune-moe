```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
array = np.array(
    [
        [0.87, 0.32, 0.38, 0.13, 0.19],
        [0.32, 0.94, 0.49, 0.08, 0.20],
        [0.38, 0.49, 0.84, 0.35, 0.10],
        [0.13, 0.08, 0.35, 0.903, 0.56],
        [0.19, 0.20, 0.10, 0.56, 0.78],
    ]
)

array = array / np.sum(array, keepdims=True, axis=1)

array
```

```python
datasets = ["RSNA-PE (CT)", "CC-CCII (CT)", "Lung-PET-CT-Dx (CT)", "CheXpert (X-ray)", "SSIM-COVID (X-ray)"]
```

```python
fig, ax = plt.subplots()
im = ax.imshow(array, cmap="Oranges")

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(datasets)), labels=datasets)
ax.set_yticks(np.arange(len(datasets)), labels=datasets)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

plt.colorbar(im)

fig.tight_layout()
plt.show()
```
